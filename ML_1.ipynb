```python
#Q1
#part h
x = np.array([[2, 3, 4], [3, 2, 9]])
y = np.array([[1, 5, 0], [5, 10, 3]])

#For matrices: the ‚Äúinner product‚Äù is often used to mean matrix multiplication (when dimensions align).
# A(2x3) x B(3x2) = C(2x2)
inner = np.inner(x,y)
print(inner)

#outer cartesian product means element wise multiplication i.e each element of matrix 1 is multiplied by corresponding element of matrix 2
#matrix size can be diff
#here axb = a x b transpose
#basically a = [1,2] , b=[3,4,5]
# step 1 : Take the first element of ùëé(which is 1) and multiply with all of b
# we get [3,4,5]
#step 2 : Take the second element of a(which is 2) and multiply with all of b
# we get [6,8,10]
#step 3 : stack them as rows and we get the answer
outer = np.outer(x,y)
print(outer)


#cartesian product : set of all ordered pairs between elements of two sets (not really a matrix operation).
#it comes from set theory
from itertools import product #intertools gives tools for creating iterators i.e things u can loop over
cartesian = np.array(list(product(x.flatten(),y.flatten())))
#after using product we get tuples like (1,3) (1,4) (2,3) (2,4)
#now we want all tuples at once so we convert them into list [(1,3),(1,4),(2,3),(2,4)]
#then we convert into arrays to make it a matrix
print(cartesian)
```

```python
#Q4
#part a

#!pip install cv2 (for jupyter notebook)
import cv2

def img_to_array(path):
  img = cv2.imread(path,cv2.IMREAD_UNCHANGED)
# keeps the image in the format given rgb or grayscale
  if img is None:
    print("Error")
    return

  if len(img.shape) == 2:
    print("Grayscale")
    arr = np.array(img)
    np.savetxt("Grayscale_image.txt",arr,fmt="%d")
    #fmt is format string i.e it controls how number in the array is written into text file
    # %d means save as integer

  elif len(img.shape) == 3 and img.shape[2] == 3:
    print("RGB")
    arr = np.array(img)
    with open("RGB_image.txt","w") as f:
      #f is file object
      #with isliye written taki file is automatically closed after writing
      for channel,color in zip(range(3),["R","G","B"]):
        #so range 2 means 0 for r
        # 1 for g
        # 2 for b
        # zip pairs numbers and channels like (0,"R") (1,"G") (2,"B")
        # on each loop channel gets number 0,1,2
        # color gets corresponding label
        f.write(f"\n{color} channel:\n") # f is formated string
        #\n means new line
        np.savetxt(f,arr[:,:,channel],fmt="%d")
        # [:,:,channel] selects one channel of rgb in 1st iteration
        # then all those pixels are converted into grayscale intensity values for that channel in form of 2d matrix
        # similarly in next iteration other channel are appended in same file
  else:
    print("Unsupported")

  print("Array saved")

img_to_array("/mountains_ML.jpeg")
```
