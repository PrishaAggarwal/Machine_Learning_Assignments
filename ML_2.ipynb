```python
#firstly downloaded data from kaggle
#part 1
#part a

import pandas as pd

df = pd.read_csv("/content/AWCustomers.csv")
print(df.head())
print(df.shape)
print(df.info())
```

```python
#since we know some typical columns like ID, names,email arent imp in database so we remove them
#we are selecting features just by analysis

df = df.drop(columns=["LastUpdated","CustomerID","Title","FirstName","MiddleName","LastName","Suffix","City","AddressLine1","AddressLine2","PostalCode","PhoneNumber"])
print(df.head())
```

```python
#Part 1
#part b

#create a new dataframe
# List of selected columns
selected_columns = [
  "StateProvinceName",
  "CountryRegionName",
  "BirthDate",
  "MaritalStatus",
  "YearlyIncome",
  "Gender",
  "TotalChildren",
  "NumberChildrenAtHome",
  "Education",
  "Occupation",
  "HomeOwnerFlag",
  "NumberCarsOwned"
]

# Create new DataFrame with only these columns
df_selected = df[selected_columns]

# Preview
print(df_selected.head())


#saved the cleaned dataset
df.to_csv("customers_cleaned.csv",index=False)
#index is false since to_csv adds by default row nnumbers onko htane ke liye
```

```python
#Part 1
#part c

#checking the type of each column
print(df.dtypes)

for col in df_selected.columns:
  print(f"{col} : {df_selected[col].unique()[:10]}")
```

```python
#hence lets determine the value type

#StateProvinceName: discrete,nominal     binning
#CountryRegionName: discerte,nominal     binning
#BirthDate: discrete,Interval            convert to age and normalizatiom
#Education: discrete,ordinal             binning
#Occupation: discrete,nominal            binning
#Gender: discrete,nominal                one-hot encoding
#MaritalStatus: discrete,nominal         one-hot encoding
#HomeOwnerFlag: discrete,nominal         one-hot encoding
#NumberCarsOwned: discrete,ratio         Normalization
#NumberChildrenAtHome: discrete,ratio    Normalization
#TotalChildren: discrete,ratio           Normalization
#YearlyIncome: continuous,ratio          Normalization

#ordinal used when category present and there is some order
#ratio used meaningful order is there with true zero
#interval used when no true zero
#nominal used when no order
```

```python
#Part 2
#lets do pre-processing for each column acc to its needs but deal with null values in each column

#first of all lets convert dob to age and remove dob from df and add age
df_selected['BirthDate'] = pd.to_datetime(df_selected['BirthDate'])
#the above code converts strings i.e "1990-05-24" to without string real time object
current_date = pd.to_datetime('today') #here we are getting todays date
df_selected['Age'] = (current_date - df_selected['BirthDate']).dt.days // 365
#in above line we got number of days after subtracting
# .dt.days diff into number of days
# // 365 does integer division to app age in years
df_selected = df_selected.drop(columns = ["BirthDate"])

print(df_selected.head())
```

```python
#Part 2
#part a

#now lets handle null values

#check for null count
print(df_selected.isnull().sum())
#hence we have no null values and lets start column wise pre-processing
```

```python
#Part 2
#part b analysis

#We have to normalize YearlyIncome,TotalChilden,NumberChildernAtHome,NumberCarsOwned,Age
#now what kind of normalization we should use ??
#Min-Max when values bounded in known range like 0-100
#Data that follows normal distribution, use z-score
#many outliers for example when income then use robust scaling or log transformation

#we have to also make sure that range/scale of each column is same

#so we can see except for YearlyIncome, we can apply min-max normalization
#for YearlyIncome we have to figure out the normalization after looking at its values
#if it has extreme values then we cant apply min-max to it and scale would vary then
#so lets figure out for income
#lets find out the mean,median and extreme values in income

mean_val = df_selected["YearlyIncome"].mean()
median_val = df_selected["YearlyIncome"].median()
max_val = df_selected["YearlyIncome"].max()
min_val = df_selected["YearlyIncome"].min()

print(mean_val,median_val,max_val,min_val)
#if mean is almost equal to median then min-max laga sakte hai since data is symmetric
#if mean >> median then data is right skewed and outliars exist

#now since mean is almost equal to median
#the diff between max and min is not extreme like 100x or 1000x
#we check outliers by finding interquartile range
Q1 = df_selected["YearlyIncome"].quantile(0.25)
Q3 = df_selected["YearlyIncome"].quantile(0.75)
IQR = Q3-Q1
Lower_Outlier = Q1 - 1.5*IQR
Upper_Outlier = Q3 + 1.5*IQR

print(Q1,Q3,IQR,Lower_Outlier,Upper_Outlier)
#hence no extreme outliers
#so we can apply min-max normalization to each column we selected for normlization


#lets check if we can apply the min-max normalization to age
mean_age = df_selected["Age"].mean()
median_age = df_selected["Age"].median()
max_age = df_selected["Age"].max()
min_age = df_selected["Age"].min()

print(mean_age,median_age,max_age,min_age)

#we also check the outliers
Q1_age = df_selected["Age"].quantile(0.25)
Q3_age = df_selected["Age"].quantile(0.75)
IQR_age = Q3_age-Q1_age
Lower_Outlier_age = Q1_age - 1.5*IQR_age
Upper_Outlier_age = Q3_age + 1.5*IQR_age

print(Q1_age,Q3_age,IQR_age,Lower_Outlier_age,Upper_Outlier_age)

#although we have outliers here and u can see outliers(77-95) to 0.75-1
#hence most data goes into 70 percent of normalized range but we dont have problem with that here so we will do min-max normalization
```

```python
#part 2
#part b

#applying min-max normalization to YearlyIncome,TotalChilden,NumberChildernAtHome,NumberCarsOnwed

from sklearn.preprocessing import MinMaxScaler

cols_to_normalize = ['YearlyIncome', 'TotalChildren', 'NumberChildrenAtHome', 'NumberCarsOwned','Age']

scaler = MinMaxScaler()
#MinMaxScalar is a class in sklearn.preprocessing
#here we are creating an object of that class and storing it in scalar
#this object now can be used to fit and transform your data

df_selected[cols_to_normalize] = scaler.fit_transform(df_selected[cols_to_normalize])
#fit_transform computes min and max for each column and scales values into[0,1]
#the original columns are replaced with normalized ones

print(df_selected.head())
```

```python
#Part 2
#part e

#since homeowner flag already is in one-hot encoding so no need to apply to  it
#we have to apply one-hot encoding to maritial status and gender

df_selected = pd.get_dummies(df_selected,columns=['MaritalStatus','Gender'],drop_first=True,dtype=int)
print(df_selected.head())

#from the above code i can see that i will get 2 features per column but since 1st feature can be infered from 2nd feature so we will avoid this dummy trap
#to avoid dummy trap we add a parameter in above code drop_first = true

#Also one recommendation, if we are doing linear/logistic regression we should use drop_first
#in case of decision tress, we should not drop that feature

#now since our output shows true/false and we want 0/1 so we add dtype=int as a paramter in above code
```

```python
#Part 3
#we have made sure that each attribute transformed is in same scale

#Part a
#now we have to apply similarity measure
#Simple Matching Coefficient for categorial or binary attributes
#Jaccard Similarity for binary attributes
#Cosine Similarity for numeric attributes

#we have been asked to pick 2 objects that means any 2 rows and apply these analysis to them
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np

print(df_selected.head(6))

#segregate columns into numerical and binary/categorical
categorial_cols = ['StateProvinceName','CountryRegionName','MaritalStatus_S','Gender_M','Education','Occupation','ContinentName','HomeOwnerFlag']
numeric_cols = ['YearlyIncome','TotalChildren','NumberChildrenAtHome','NumberCarsOwned','Age']
binary_cols = ['MaritalStatus_S','Gender_M','HomeOwnerFlag']

#lets take 2 rows from dataset [0] and [5]
#loc is used to access rows and columns by labels
# .values is used to convert dataframe/series into numpy array
obj1 = df_selected.loc[0,categorial_cols].values
obj2 = df_selected.loc[5,categorial_cols].values

#finding SMC for the state,country,occupation,education
#SMC = number of matching attributes/total attributes
smc = np.mean(obj1 == obj2)
print(smc)

#finding jaccard similarity
#it only works in 0/1
obj3 = df_selected.loc[0,binary_cols].values
obj4 = df_selected.loc[5,binary_cols].values
intersection = np.sum(np.minimum(obj3,obj4)) #number of positions in which both have 1
#np. minimum gives element wise minimum
#min(1,1) = 1
#min(1,0) = 0
#min(0,1) = 0
#min(0,0) = 0
#summing over min gives intersection
union = np.sum(np.maximum(obj3,obj4)) #number of positions where at least one has 1
#max(1,1) = 1
#max(1,0) = 1
jaccard= intersection/union
print(jaccard)


obj5 = df_selected.loc[0,numeric_cols].values
obj6 = df_selected.loc[5,numeric_cols].values

cos_sim = cosine_similarity([obj5], [obj6])[0][0]
#now since we are comparing 1 row with 1 row it returns 2d array
#[0][0]  makes sure we get 1st element in 1st row of that 2d array
print(cos_sim)
```

```python
#Part 3
#part b

#in the dataset given to us we only have yearlyincome and not Commute Distance
#so lets generate distance for all rows and then find correlation between 2 features

#lets generate distance based on occupation
compute_range = {
    "Clerical" : (1,10), #means iss occupation ka distance mostly 1-10 km hoga
    "Skilled Manual" : (5,20),
    "Professional" : (10,40),
    "Management" : (5,25)
}

np.random.seed(45) #everytime same random numbers are used

#lets commute distance within assigned range
def generate_commute(occupation):
  if occupation in compute_range:
    low,high = commute_range[occupation]
    return np.random.uniform(low,high+1) #+1 so that high wala number is included
  else:
    return np.random.uniform(1,10) #randint generates random integers between given numbers range
#now but i want float values as well so i use uniform
df_selected["CommuteDistance"] = df_selected["Occupation"].apply(generate_commute)
print(df_selected.head())

#now since we have generated commute distance and we have yearly income
#lets find the correlation between both

correlation = df_selected["YearlyIncome"].corr(df_selected["CommuteDistance"])
print(correlation)
#+1 means strong positive correlation
#0 means no linear relationship
#-1 means negative correlation
```
